# 前言

数据源是来自哔哩哔哩UP主戴戴戴师兄的 2 份餐饮店外卖订单数据集，分成 2 部分，每部分只对其中的一个数据集进行分析。


## part 1

orders.csv 这个数据集是订单数据，可供挖掘的纬度有订单时间、配送地址，指标则有菜品个数、商家实收、商家补贴、平台补贴、订单数、订单金额、地区分布、下单平台、下单时间趋势等。

### 1、数据理解与处理

```python
import pandas as pd

data = pd.read_csv('orders.csv',encoding='gbk')
data.head()
```

```python
data.info()  # 数据集情况 4419 条，37个字段
```

```python
data.columns = data.columns.str.strip()  # 列名有空格，需要处理下
data.columns
```

```python
data[data.duplicated()].count()  # 没有完全重复的数据
```

```python
data.isnull().sum()   # order_15d等存在空值，表示有多少用户在15天内未下单的
```

```python
data.rename(columns={'配送地址eleme':'配送地址'},inplace=True) # 修改列名
```

### 2、数据分析可视化

#### 2.1 整体情况

```python
result = {}
result['总订单数'] = data['订单id'].count()  
result['已完成订单数'] = data['订单id'][data['下单时间'].notnull()].count()  
result['未付款订单数'] = data['订单id'][data['下单时间'].isnull()].count()   
result['总订单金额'] = data['订单金额'][data['下单时间'].notnull()].sum()  
result['用户实付金额'] = data['用户实付'][data['下单时间'].notnull()].sum()
result['商家补贴总金额'] = data['商家补贴'][data['下单时间'].notnull()].sum()
result['平台补贴总金额'] = data['平台补贴'][data['下单时间'].notnull()].sum()
result['总补贴金额'] = data['总补贴'][data['下单时间'].notnull()].sum()
result['商家总实收金额'] = data['商家实收'][data['下单时间'].notnull()].sum()
```

```python
result
```

```python
from pyecharts import options as opts
from pyecharts.charts import Map, Bar, Line
from pyecharts.components import Table
from pyecharts.options import ComponentTitleOpts
from pyecharts.faker import Faker

table = Table()

headers = ['总订单数', '总订单金额','已完成订单数', '未付款订单数','用户实付金额', '商家总实收金额', '总补贴金额', '成交率']
rows = [
    [
        result['总订单数'], f"{result['总订单金额']/10000:.2f} 万", result['已完成订单数'],result['未付款订单数'],
        f"{result['用户实付金额']/10000:.2f} 万", 
        f"{result['商家总实收金额']/10000:.2f} 万",
        f"{result['总补贴金额']/10000:.2f} 万", 
        f"{result['已完成订单数']/result['总订单数']:.2%}",
    ]
]
table.add(headers, rows)
table.set_global_opts(
    title_opts=ComponentTitleOpts(title='整体情况')
)
table.render_notebook()
```

#### 2.2 配送地区分析

```python
result2 = data[data['下单日期时间'].notnull()].groupby('配送地址').agg({'订单id':'count'})
result21 = result2.to_dict()['订单id']
c = (
    Map()
    .add("订单量", [*result21.items()], "上海",is_map_symbol_show=False)
    .set_series_opts(label_opts=opts.LabelOpts(is_show=True))
    .set_global_opts(
        title_opts=opts.TitleOpts(title='配送地区分布'),
        visualmap_opts=opts.VisualMapOpts(max_=1000),            
    )
)
c.render_notebook()
```

#### 2.3 时间分析

```python
data['下单日期时间'] = pd.to_datetime(data['下单日期时间'])
```

```python
result31 = data.groupby(data['下单日期时间'].apply(lambda x: x.strftime("%Y-%m-%d"))).agg({'订单id':'count'}).to_dict()['订单id']
c = (
    Line()
    .add_xaxis(list(result31.keys()))
    .add_yaxis("订单量", list(result31.values()))
    .set_series_opts(
        label_opts=opts.LabelOpts(is_show=False),
        markpoint_opts=opts.MarkPointOpts(
            data=[
                opts.MarkPointItem(type_="max", name="最大值"),
            ]
        ),
    )
    .set_global_opts(title_opts=opts.TitleOpts(title="每日订单量走势"))
)
c.render_notebook()
```

从上图来看，8月份上半月订单量比较少，下半月的订单量增长明显。

```python
result32 = data.groupby(data['下单日期时间'].apply(lambda x: x.strftime("%H"))).agg({'订单id':'count'}).to_dict()['订单id']
x = [*result32.keys()]
y = [*result32.values()]
c = (
    Bar()
    .add_xaxis(x)
    .add_yaxis("订单量", y)
    .set_global_opts(title_opts=opts.TitleOpts(title="每小时订单量走势"))
    .set_series_opts(
        label_opts=opts.LabelOpts(is_show=False),
        markpoint_opts=opts.MarkPointOpts(
            data=[
                opts.MarkPointItem(type_="max", name="峰值"),
                opts.MarkPointItem(name="第二峰值", coord=[x[7], y[7]], value=y[7]),
                opts.MarkPointItem(name="第三峰值", coord=[x[5], y[5]], value=y[5]),
            ]
        ),
    )
)
c.render_notebook()
```

从每小时订单量走势来看，一天中有3个高峰期（10点、11点、12点），其中11点-12点之间是一天中订单量最多的时候。对于卖家的指导意义就是，为了提高订单量，高峰期时应该尽量保证客服的回复速度，尤其是中午11点-12点之间.

#### 2.1 每日整体门店营业额走势

```python
result = data2.groupby('日期').agg({'门店营业额':'sum'}).to_dict()['门店营业额']
c = (
    Line()
    .add_xaxis(list(result.keys()))
    .add_yaxis("营业额", list(result.values()))
    .set_series_opts(
        areastyle_opts=opts.AreaStyleOpts(opacity=0.5),
        label_opts=opts.LabelOpts(is_show=False),
        markpoint_opts=opts.MarkPointOpts(
            data=[
                opts.MarkPointItem(type_="max", name="最大值"),
                opts.MarkPointItem(type_="min", name="最小值"),
                opts.MarkPointItem(type_="average", name="平均值"),
            ]
        ),
    )
    .set_global_opts(title_opts=opts.TitleOpts(title="每日整体门店营业额走势"))
)
c.render_notebook()
```

#### 2.2 哪家门店营业额最高

```python
dts = list(data2['日期'].unique())
dts.reverse()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Map, Timeline, Bar, Line, Pie
from pyecharts.components import Table
from pyecharts.options import ComponentTitleOpts

tl = Timeline()
tl.add_schema(
        is_auto_play=True,
        is_loop_play=False,
        play_interval=500,
    )
for dt in dts:
    item = data2[data2['日期'] <= dt].groupby('平台门店名称').agg({'门店下单量': 'sum', '门店营业额': 'sum'}).sort_values(by='门店下单量', ascending=False)[:10].sort_values(by='门店下单量').to_dict()
    bar = (
        Bar()
        .add_xaxis([*item['门店下单量'].keys()])
        .add_yaxis("下单量", [val for val in item['门店下单量'].values()], label_opts=opts.LabelOpts(position="right"))
        .add_yaxis("营业额", [val for val in item['门店营业额'].values()], label_opts=opts.LabelOpts(position="right"))
        .reversal_axis()
        .set_global_opts(title_opts=opts.TitleOpts(title="累计下单量排行 TOP10")
        )
    )
    tl.add(bar, dt)
tl.render_notebook()
```

```python
item = data2.groupby('平台门店名称').agg({'门店下单量': 'sum'}).sort_values(by='门店下单量', ascending=False)[:10].to_dict()['门店下单量']
item = {k: v for k, v in item.items()}
c = (
    Pie()
    .add("下单量", [*item.items()])
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c} ({d}%)"))
)
c.render_notebook()
```
